# 04 Creating a spending flag

#06. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:

-If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”

-If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# View relevant columns 
ords_prods_merge[['user_id', 'mean_product_price']].head(20)
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(20)
```

# 05 Creating a order frequency flag

```python
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'median_days_since_prior_order']].head(20)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] <= 20) & (ords_prods_merge['median_days_since_prior_order'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 20, 'order_frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merge[['user_id', 'median_days_since_prior_order', 'order_frequency_flag']].head(20)
```
